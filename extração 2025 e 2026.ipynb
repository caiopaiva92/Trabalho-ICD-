```python
meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho',
         'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
```

```python
def csv_creator(links_meses, ano, meses):
    i = 0
    ids_categorias = [('osversion_details', 'OS Version'), ('cat0_details', 'System RAM'),
                      ('cat2_details', 'Physical CPUs'), ('cat3_details', 'Video Card Description'),
                      ('cat4_details', 'VRAM'), ('cat5_details', 'Primary Display Resolution')]

    with open(f'dados_mensais_{ano}.csv', 'w', encoding='utf-8') as f:
        f.write('Categoria,Especificação,Porcentagens,Incrementos/Decrementos,Mês,Ano\n')

    for link in links_meses:
        time.sleep(5)
        html = requests.get(link).text
        soup = BeautifulSoup(html, 'html.parser')

        for id_, categoria in ids_categorias:
            div = soup.find('div', id=f'{id_}')
            titulos = div.find_all('div', class_='stats_col_mid')
            porcentagens = div.find_all('div', class_='stats_col_right')
            incs_decs = div.find_all('div', class_='stats_col_right2')

            with open(f'dados_mensais_{ano}.csv', 'a+', encoding='utf-8') as f:
                for titulo, porcentagem, inc_dec in zip(titulos, porcentagens, incs_decs):
                    f.write(f"{categoria},{titulo.text.strip()},{porcentagem.text.strip()},{inc_dec.text.strip()},{meses[i]},{ano}\n")
        i += 1
```

```python
csv_creator(link_mensal_2025, 2025, meses)
```

```python
csv_creator(link_mensal_2026, 2026, meses[:5])
```

# Criação dos dados anuais de 2025

```python
import pandas as pd

ano = 2025
df = pd.read_csv(f'dados_mensais_{ano}.csv')

# Conversão de porcentagem e incremento para float
df['Porcentagens'] = df['Porcentagens'].str.replace('%', '', regex=False).str.strip().astype(float)
df['Incrementos/Decrementos'] = df['Incrementos/Decrementos'].str.replace('%', '', regex=False).str.strip().astype(float)

# Valores de Dezembro como referência final do ano
dezembro = (df[df['Mês'] == 'Dezembro']
            [['Ano', 'Categoria', 'Especificação', 'Porcentagens']].copy())

# Primeira aparição de cada especificação ao longo do ano
primeira = (df.groupby(['Ano', 'Categoria', 'Especificação'], as_index=False).first()
            [['Ano', 'Categoria', 'Especificação', 'Porcentagens', 'Mês']]
            .rename(columns={'Porcentagens': 'Porcentagem_Inicial', 'Mês': 'Primeiro_Mês'}))

# União e cálculo da variação anual (Dezembro - primeira aparição)
resultado = pd.merge(dezembro, primeira, on=['Ano', 'Categoria', 'Especificação'], how='inner')
resultado['Incrementos/Decrementos'] = (resultado['Porcentagens'] - resultado['Porcentagem_Inicial']).round(2)

print(f"Registros anuais: {len(resultado)}")
resultado.head(10)
```

```python
csv_final = resultado[['Categoria', 'Especificação', 'Porcentagens', 'Incrementos/Decrementos', 'Ano']]
csv_final.to_csv('dados_anuais_2025.csv', index=False)
```
